# Import Data

```python
GP = pd.read_csv('https://raw.githubusercontent.com/YiyangZhangyz738/Gas-Price/main/data/Queried_Data/GP.csv',parse_dates=['Date(UTC)'])
GP.head()
```

# Process Data

```python
column_2 = []
column_4 = []

column_2 = GP.iloc[:, 1]
column_4 = GP.iloc[:, 3]

change_rates = []
change_rates.append(0)
for i in range(1, len(column_4)):
  if column_4[i-1] == 0:
    change_rates.append(0)
  else:
    change_rate = (float(column_4[i]) - float(column_4[i-1])) / float(column_4[i-1])
    change_rates.append(change_rate)

rate = {}
GPR = pd.DataFrame(rate)
GPR['Date']=column_2
GPR['Rate']=change_rates
GPR=GPR.set_index(GPR.columns[0])
GPR.head()
```

# Export Data

```python
GPR.to_csv("GPR.csv")
```
